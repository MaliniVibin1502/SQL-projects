List category name and description from the Categories table

```python
%%sql
SELECT categoryname, description
FROM categories
ORDER BY categoryname
```

List name, address, city of all customers which are not from 'Germany', 'Mexico', 'Spain'

```python
%%sql
SELECT contactname,address,city
FROM customers
WHERE country NOT IN ("Germany","Mexico","Spain")
ORDER BY city
LIMIT 100
```

List order date, shipped date, customer id, Freight of all orders placed on 2018 Feb 26

```python
%%sql
SELECT orderdate, shippeddate,customerid,freight
FROM orders
WHERE orderdate = "2018-02-26"
```


List employee_id, order_id, customer_id, required_date, shipped_date from all orders shipped later than the required date

```python
%%sql
SELECT employeeid,orderid,customerid,requireddate,shippeddate,requireddate
FROM orders
WHERE shippeddate > requireddate
ORDER BY shippeddate
LIMIT 100
```

Listing all the even numbered Order_id from the orders table

```python
%%sql
SELECT *
FROM orders
WHERE orderid % 2 = 0
ORDER BY orderid
LIMIT 100
```

List city, company_name, contact_name of all customers from cities which contains the letter 'L' in the city name, sorted by contact_name

```python
%%sql
SELECT companyname, contactname, city
FROM customers
WHERE city LIKE '%L%'
ORDER BY city
LIMIT 50
```

List company_name, contact_name, fax number of all customers that has a fax number. (not null)

```python
%%sql
SELECT companyname,contactname,fax
FROM customers
WHERE fax IS NOT NULL
LIMIT 50
```

List first_name, last_name. hire_date of the most recently hired employee.

```python
%%sql
SELECT firstname,lastname,MAX(hiredate) AS hiredate
FROM employees
ORDER BY firstname
```

Show the average unit price rounded to 2 decimal places, the total units in stock, total discontinued products from the products table.

```python
%%sql
SELECT ROUND(AVG(unitprice),2)  AS Unitpirce_Avg,SUM(unitsinstock) AS Total_stock,SUM(discontinued) AS Total_discontinued
FROM products

```

Show the ProductName, CompanyName, CategoryName from the products, suppliers, and categories table

```python
%%sql
SELECT products.productname, categories.categoryname,suppliers.companyname
FROM products
JOIN categories ON categories.categoryid = products.categoryid
JOIN suppliers ON suppliers.supplierid = products.supplierid
ORDER BY productname
LIMIT 20
```

Show the category name and the average product unit price for each category rounded to 2 decimal places.

```python
%%sql
SELECT categories.categoryname, products.productname, ROUND(AVG(products.unitprice)) AS Avg_unitprice
FROM products
JOIN categories ON categories.categoryid = products.categoryid
GROUP BY categoryname
LIMIT 50
```
