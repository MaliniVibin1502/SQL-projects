Retrieve the distribution of test scores (i.e. how many students got each score). Assign the result of this query to the variable `test_score_distribution`

```python
test_score_distribution= %sql SELECT test_score, COUNT(test_score) as TOTAL FROM INTERNATIONAL_STUDENT_TEST_SCORES GROUP BY test_score
test_score_distribution
```
